```python
# Ask the user which country they desire to travel to.

while True:
    try:
        country_code = input('Please enter an ISO 3166 code for the country to which you would like to travel: ').upper()
    except ValueError:
        print('Please enter an ISO 3166 country code.')
        print(32)

    if country_code not in country_tuple:
        print('Country code not in list. Please try again.')
        print(35)
        continue
    else:
        break
```

```python
# Filter for selected country

selected_country_df = weatherpy_vacation_df.loc[(weatherpy_vacation_df['Country']==country_code)]

# Randomly select a number of cities
n = 5
city_sample_df = selected_country_df.sample(n)

city_sample_df.reset_index(drop=True, inplace=True)
city_sample_df.info()
```

```python
# Get cities' lattitude and longitude pairs
# https://stackoverflow.com/questions/16729574/how-to-get-a-value-from-a-cell-of-a-dataframe

city_1_lat = city_sample_df.iloc[0]['Lat']
city_1_lng = city_sample_df.iloc[0]['Lng']
city_2_lat = city_sample_df.iloc[1]['Lat']
city_2_lng = city_sample_df.iloc[1]['Lng']
city_3_lat = city_sample_df.iloc[2]['Lat']
city_3_lng = city_sample_df.iloc[2]['Lng']
city_4_lat = city_sample_df.iloc[3]['Lat']
city_4_lng = city_sample_df.iloc[3]['Lng']
city_5_lat = city_sample_df.iloc[4]['Lat']
city_5_lng = city_sample_df.iloc[4]['Lng']

city_1 = (city_1_lat,city_1_lng)
city_2 = (city_2_lat,city_2_lng)
city_3 = (city_3_lat,city_3_lng)
city_4 = (city_4_lat,city_4_lng)
city_5 = (city_5_lat,city_5_lng)
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in city_sample_df.iterrows()]

# Gmaps directions documentation
# https://jupyter-gmaps.readthedocs.io/en/latest/tutorial.html#directions-layer
fig = gmaps.figure()
locations = city_sample_df[['Lat', 'Lng']]
max_temp = city_sample_df['Max Temp']
directions = gmaps.directions_layer(city_1, city_2, show_markers=False,waypoints=[city_3,city_4,city_5],
                                    travel_mode='DRIVING')
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig.add_layer(directions)
fig
```
